# Creating Visualization using Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Ice Cream Ratings.csv", index_col="Date")
```

```python
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-deep')
```

#### df.plot()

```python
df.plot()
```

```python
df.plot(kind='line', title="Ice Cream Ratings", xlabel='Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind='line', subplots=True)
```

```python
df["Flavor Rating"].plot()
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot(kind='barh', stacked=True)
```

```python
df.plot(kind='scatter', x='Texture Rating', y='Overall Rating', s=500, c='r')
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot(kind='area',figsize=(10,5))
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize=(10,10))
```
